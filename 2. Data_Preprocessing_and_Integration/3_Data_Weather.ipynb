```python
import pandas as pd
```

```python
weather_df = pd.read_csv(r"D:\Data_D\TMU\MRP\1.0.Dataset\Weather\4301028.csv")

weather_df.head()
```

```python
weather_df = weather_df.rename(columns={'DATE':'date'})
weather_df['date'] = pd.to_datetime(weather_df['date'])
weather_df.head(1)
```

```python
weather_df = weather_df[['date', 'TMAX', 'TMIN', 'PRCP', 'SNOW', 'AWND']]
weather_df.head(1)
```

```python
weather_df.shape
```

```python
weather_df[['TMAX', 'TMIN', 'PRCP', 'SNOW', 'AWND']].isnull().sum()
```

```python
weather_df['temp_avg'] = (weather_df['TMAX'] + weather_df['TMIN'])/2
weather_df['is_rain'] = (weather_df['PRCP']>0).astype(int)
weather_df['is_snow'] = (weather_df['SNOW']>0).astype(int)
```

```python
def temp_category(temp):
    if temp < 32:
        return 'cold'
    elif temp < 60:
        return 'mild'
    else:
        return 'hot'
```

```python
weather_df['temp_category'] = weather_df['temp_avg'].apply(temp_category)
```

```python
weather_df['high_wind'] = (weather_df['AWND']>15).astype(int)
```

```python
weather_df.to_csv(r"D:\Data_D\TMU\MRP\1.0.Dataset\Weather\Weather_Final.csv")
```
