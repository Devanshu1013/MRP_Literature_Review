```python
import pandas as pd
```

```python
# Load Final Crime + Socio + Weather Dataset
crime_final = pd.read_csv(
    r"D:\Data_D\TMU\MRP\1.0.Dataset\Crime_Socio_Economic_Weather.csv"
)

# Load Mobility Dataset
mobility = pd.read_csv(
    r"D:\Data_D\TMU\MRP\1.0.Dataset/Mobality/Mobality_Final.csv"
)
```

```python
crime_final['date'] = pd.to_datetime(crime_final['date']).dt.date
mobility['date'] = pd.to_datetime(mobility['date']).dt.date
```

```python
print("Crime Dataset Shape :", crime_final.shape)
print("Mobility Dataset Shape :", mobility.shape)
```

```python
final_df = pd.merge(
    crime_final,
    mobility,
    on='date',
    how='left'
)
```

```python
print("Final Dataset Shape :", final_df.shape)
```

```python
print(final_df.isnull().sum())
```
