```python
crime_df = pd.read_csv(r"D:\Data_D\TMU\MRP\1.0.Dataset\Crime\Crime_2001_2023_Cleaned.csv")
socio_final = pd.read_csv(r"D:\Data_D\TMU\MRP\1.0.Dataset\Socio_Economic\Socio_Economic_Final.csv")
crime_df.columns
```

```python
# Rename for consistency
crime_df = crime_df.rename(columns={
    'Community Area': 'community_area',
    'Year': 'year'
})
```

```python
# Period 1 → use 2012 socio data
period_1 = crime_df[crime_df['year'] <= 2012].copy()

# Period 2 → use 2014 socio data
period_2 = crime_df[(crime_df['year'] >= 2013) & (crime_df['year'] <= 2016)].copy()

# Period 3 → use 2020 socio data
period_3 = crime_df[crime_df['year'] >= 2017].copy()
```

```python
socio_2012 = socio_final[socio_final['year'] == 2012].copy()
socio_2014 = socio_final[socio_final['year'] == 2014].copy()
socio_2020 = socio_final[socio_final['year'] == 2020].copy()
```

```python
socio_2012 = socio_2012.drop(columns=['year'])
socio_2014 = socio_2014.drop(columns=['year'])
socio_2020 = socio_2020.drop(columns=['year'])
```

```python
merged_1 = period_1.merge(socio_2012, on='community_area', how='left')
merged_2 = period_2.merge(socio_2014, on='community_area', how='left')
merged_3 = period_3.merge(socio_2020, on='community_area', how='left')
```

```python
final_merged_df = pd.concat([merged_1, merged_2, merged_3], ignore_index=True)

final_merged_df.head()
```

```python
final_merged_df[['per_capita_income', 'unemployment_rate', 'no_highschool_pct']].isnull().sum()
```

# Weather Data

```python
weather_df = pd.read_csv(r"D:\Data_D\TMU\MRP\1.0.Dataset\Weather\4301028.csv")

weather_df.head()
```

```python
weather_df = weather_df.rename(columns={'DATE': 'date'})

weather_df['date'] = pd.to_datetime(weather_df['date'])

weather_df.head()
```

```python
final_merged_df['date'] = pd.to_datetime(final_merged_df['Date']).dt.date

final_merged_df['date'] = pd.to_datetime(final_merged_df['date'])

final_merged_df[['Date', 'date']].head()
```

```python
final_merged_df = final_merged_df.merge(
    weather_df[['date', 'TMAX', 'TMIN', 'PRCP', 'SNOW', 'AWND']],
    on='date',
    how='left'
)

final_merged_df.head()
```

```python
final_merged_df[['TMAX', 'TMIN', 'PRCP', 'SNOW', 'AWND']].isnull().sum()
```

```python
final_merged_df['temp_avg'] = (final_merged_df['TMAX'] + final_merged_df['TMIN']) / 2
```

```python
final_merged_df['is_rain'] = (final_merged_df['PRCP'] > 0).astype(int)
```

```python
final_merged_df['is_snow'] = (final_merged_df['SNOW'] > 0).astype(int)
```

```python
def temp_category(temp):
    if temp < 32:
        return 'cold'
    elif temp < 60:
        return 'mild'
    else:
        return 'hot'

final_merged_df['temp_category'] = final_merged_df['temp_avg'].apply(temp_category)
```

```python
final_merged_df['high_wind'] = (final_merged_df['AWND'] > 15).astype(int)
```

```python
final_merged_df.to_csv(
    r"D:\Data_D\TMU\MRP\1.0.Dataset\Final_Crime_Full_Features.csv",
    index=False
)
```
